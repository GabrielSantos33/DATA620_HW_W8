```python
emma = nltk.corpus.gutenberg.words('austen-emma.txt')

emma = [w for w in emma if w.isalpha()]

from nltk.corpus import stopwords
emma = [w for w in emma if w not in stopwords.words('english')]
```

4.Identify the 200 highest frequency words in this corpus.

I am going to identify the 200 words with the highest frequency words in this corpus.

```python
freq_emma = nltk.FreqDist(emma)

emma_top200 = pd.DataFrame(freq_emma.most_common(200), columns = ['word','count'])
emma_top200.head(15)
```

The first three common mass words are: I (3178 times), Mr (1153 times) and Emma (865 times).

```python
emma_top500 = pd.DataFrame(freq_emma.most_common(500),columns=['word','count'])
```

3.Taking the most common words, how many unique words represent half of the total words in the corpus?

Taking the most common words. I found that the 231 most common words make up 50% of all words in the corpus.

```python
counter = 0
half_total = []
for word in freq_emma.most_common():
    if counter >= len(emma)/2:
        break
    else:
        counter += word[1]
        half_total.append(word)
        
len(half_total)
```

6.Does the observed relative frequency of these words follow Zipfs law? Explain.

Zipf's Law suggests that the most frequent word will appear approximately twice as often as the second most frequent word, three times as often as the third most frequent word, etc. However, that doesn't seem to be the case here.

The first four most common words are: I (3178 times), Mr (1153 times), Emma (865 times), could (825 times).

Looking at the first four most common words we can see that it does not follow Zipf's law.
